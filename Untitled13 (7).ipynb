```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##### TAKING DATASETS #####

```python
df = pd.read_csv('/content/water-level_turbidity-high.csv.zip')
```

```python
for col in df.columns:
    if 'date' in col.lower() or 'time' in col.lower():
        df[col] = pd.to_datetime(df[col])
        time_col = col
        break
else:
    time_col = None
```

```python
possible_level_cols = ['water_level', 'level', 'tank_level', 'waterlevel']
level_col = None

for col in df.columns:
    if col.lower() in possible_level_cols:
        level_col = col
        break

# If not auto-detected, select numeric column
if level_col is None:
    numeric_cols = df.select_dtypes(include=np.number).columns
    level_col = numeric_cols[0]

print(f"\nUsing '{level_col}' as Water Level column")
```

```python
plt.figure(figsize=(12, 5))

if time_col:
    plt.plot(df[time_col], df[level_col])
    plt.xlabel("Time")
else:
    plt.plot(df[level_col])
    plt.xlabel("Readings")

plt.ylabel("Water Level")
plt.title("Water Tank Level Over Time")
plt.grid(True)
plt.show()
```

```python
LOW_LEVEL_THRESHOLD = df[level_col].quantile(0.25)

df['Refill_Required'] = df[level_col] < LOW_LEVEL_THRESHOLD

print("\nLow Water Threshold:", LOW_LEVEL_THRESHOLD)

print("\nRefill Required Samples:")
display(df[df['Refill_Required']].head())
```

```python
refill_count = df['Refill_Required'].sum()
print(f"\nTotal Refill Alerts: {refill_count}")
```

```python
plt.figure(figsize=(12, 5))

plt.plot(df[level_col], label="Water Level")
plt.axhline(LOW_LEVEL_THRESHOLD, linestyle='--', label="Low Level Threshold")

alert_indices = df[df['Refill_Required']].index
plt.scatter(alert_indices, df.loc[alert_indices, level_col], color='red', label="Refill Alert")

plt.xlabel("Readings")
plt.ylabel("Water Level")
plt.title("Water Level with Refill Alerts")
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("\nProject Summary:")
print(f"Average Water Level: {df[level_col].mean():.2f}")
print(f"Minimum Water Level: {df[level_col].min():.2f}")
print(f"Maximum Water Level: {df[level_col].max():.2f}")
print(f"Refill Alerts Generated: {refill_count}")
```
